# 🏀 NBA Player Points — Exploratory Data Analysis

This notebook explores the NBA game log data before modeling. We look at:
- Distribution of points scored
- Relationships between key stats
- How rolling averages relate to actual performance
- Home vs. away and rest-day effects

```python
import sys, os
sys.path.insert(0, os.path.join('..', 'src'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from data_loader import load_data
from features import build_features, clean_raw_data

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)
```

## 1. Load the data

```python
raw = load_data(use_cache=True)
df = clean_raw_data(raw)
print(f'Shape: {df.shape}')
df.head()
```

## 2. Distribution of points scored

```python
plt.figure()
sns.histplot(df['PTS'], bins=30, kde=True, color='steelblue')
plt.title('Distribution of Points Scored')
plt.xlabel('Points')
plt.show()

print(df['PTS'].describe())
```

## 3. Points by player

```python
plt.figure(figsize=(10, 6))
order = df.groupby('PLAYER_NAME')['PTS'].median().sort_values(ascending=False).index
sns.boxplot(data=df, x='PTS', y='PLAYER_NAME', order=order, palette='Blues_r')
plt.title('Points Distribution by Player')
plt.show()
```

## 4. Correlation heatmap

```python
numeric_cols = ['PTS', 'REB', 'AST', 'MIN', 'FGA', 'FG_PCT']
corr = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Between Key Stats')
plt.show()
```

## 5. Engineered features preview

Check the feature matrix that feeds the model.

```python
X, y = build_features(raw)
X.head()
```

```python
# Does last-5-game scoring average track actual points?
plt.figure()
sns.scatterplot(x=X['rolling_pts_5'], y=y, alpha=0.4, color='steelblue')
plt.title('Rolling 5-Game Avg Points vs. Actual Points')
plt.xlabel('Rolling 5-game avg points')
plt.ylabel('Actual points')
plt.show()
```
